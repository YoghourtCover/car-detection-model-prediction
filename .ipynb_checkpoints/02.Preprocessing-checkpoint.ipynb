# Add Zeros

We use this functions to make any shape of image to become square shape with any desired one. It add zeros in such a way that image become square shape by keeping the scale ratio same. Then resize it to the desired square shape.

```python
def resizeAndPad(img, size, padColor=0):

    h, w = img.shape[:2]
    sh, sw = size

    # interpolation method
    if h > sh or w > sw: # shrinking image
        interp = cv2.INTER_AREA
    else: # stretching image
        interp = cv2.INTER_CUBIC

    # aspect ratio of image
    aspect = w/h

    # compute scaling and pad sizing
    if aspect > 1: # horizontal image
        new_w = sw
        new_h = np.round(new_w/aspect).astype(int)
        pad_vert = (sh-new_h)/2
        pad_top, pad_bot = np.floor(pad_vert).astype(int), np.ceil(pad_vert).astype(int)
        pad_left, pad_right = 0, 0
    elif aspect < 1: # vertical image
        new_h = sh
        new_w = np.round(new_h*aspect).astype(int)
        pad_horz = (sw-new_w)/2
        pad_left, pad_right = np.floor(pad_horz).astype(int), np.ceil(pad_horz).astype(int)
        pad_top, pad_bot = 0, 0
    else: # square image
        new_h, new_w = sh, sw
        pad_left, pad_right, pad_top, pad_bot = 0, 0, 0, 0

    # set pad color
    if len(img.shape) is 3 and not isinstance(padColor, (list, tuple, np.ndarray)): # color image but only one color provided
        padColor = [padColor]*3

    # scale and pad
    scaled_img = cv2.resize(img, (new_w, new_h), interpolation=interp)
    scaled_img = cv2.copyMakeBorder(scaled_img, pad_top, pad_bot, pad_left, pad_right, borderType=cv2.BORDER_CONSTANT, value=padColor)
    
    return scaled_img
```

# Preprocessing Function

```python
WHITE = [255,255,255]

details = {}

details['hdf5_dataset_label'] = ['batch', 'label', 'hotlabel', 'img_name']
details['read_path'] = 'read/it/from/this/path/'
details['save_path'] = 'save/to/this/path/'
details['resized_img_size'] = (300, 300, 3)

details['image_path'] = ['VW-Passat', 'RENO-Fluence', 'FIAT-Linea', 'VW-Polo',\
                              'RENO-Toros', 'FIAT-Dogan', 'OtherClass']
details['image_label'] = ['passat', 'fluence', 'linea', 'polo', 'toros', 'dogan', 'other']

details['read_prefix'] = 'orig_'

details['save_prefix'] = 'preprocessed_'
```

```python
# funsction for unmasking and resizing
def original2preprocessed(details):
    
    # loop over train addresses

    for class_name in details['image_path']:

        hdf5_path = details['read_path'] + details['read_prefix'] + class_name + '.hdf5'
       
        # opening like this, solves the problem of 'Unable to truncate a file which is already open'
        with h5py.File(hdf5_path,'r') as hdf5_file:
            
            batch_name_read = hdf5_file[details['hdf5_dataset_label'][0]]
            label_name_read = hdf5_file[details['hdf5_dataset_label'][1]]
            image_name_read = hdf5_file[details['hdf5_dataset_label'][3]]

            
            hdf5_path = details['save_path'] + details['save_prefix'] + class_name + '.hdf5'
            
            with h5py.File(hdf5_path,'w') as hdf5_file:
                
                data_shape = (len(batch_name_read), details['resized_img_size'][0], \
                              details['resized_img_size'][1], details['resized_img_size'][2])

                hdf5_file.create_dataset(details['hdf5_dataset_label'][0], data_shape, np.uint8)

                hdf5_file.create_dataset(details['hdf5_dataset_label'][1], (len(label_name_read), ), np.uint8)
                hdf5_file[details['hdf5_dataset_label'][1]][...] = label_name_read
                
                dt = h5py.special_dtype(vlen=bytes)
                hdf5_file.create_dataset(details['hdf5_dataset_label'][3], (len(image_name_read), ), dtype=dt)
                hdf5_file[details['hdf5_dataset_label'][3]][...] = image_name_read
                
                for index, img in enumerate(batch_name_read):

                    #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    
                    # Inpainting operation
                    img = resizeAndPad(img, (details['resized_img_size'][0], \
                                            details['resized_img_size'][1]), padColor=0)

                    # save the image
                    hdf5_file[details['hdf5_dataset_label'][0]][index, ...] = img[None]
                    # save the meanand close the hdf5 file
```

```python
original2preprocessed(details)
```
